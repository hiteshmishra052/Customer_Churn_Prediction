```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/customer_churn.csv')
df.head()
```

```python
df1 = df.drop('customerID', axis='columns')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
def fix_totalcharges(x):
  try:
    return float(x)
  except:
    return np.nan
```

```python
df1['TotalCharges'] = df1['TotalCharges'].apply(fix_totalcharges)
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df2 = df1.dropna(axis='rows')
df2.head()
```

```python
df2.dtypes
```

```python
non_num = []
for col in df2.columns:
  if(df2[col].dtype == object):
    non_num.append(col)
```

```python
print(non_num)
```

```python
for cols in non_num:
  print(cols, df2[cols].unique())
```

```python
def fix_yes_no(x):
  if(x == 'Yes'):
    return 'Yes'
  else:
    return 'No'
```

```python
df2['MultipleLines'] = df2['MultipleLines'].apply(fix_yes_no)
df2['OnlineSecurity'] = df2['OnlineSecurity'].apply(fix_yes_no)
df2['OnlineBackup'] = df2['OnlineBackup'].apply(fix_yes_no)
df2['DeviceProtection'] = df2['DeviceProtection'].apply(fix_yes_no)
df2['TechSupport'] = df2['TechSupport'].apply(fix_yes_no)
df2['StreamingTV'] = df2['StreamingTV'].apply(fix_yes_no)
df2['StreamingMovies'] = df2['StreamingMovies'].apply(fix_yes_no)
```

```python
for cols in non_num:
  print(cols, df2[cols].unique())
```

```python
for cols in non_num:
  print(cols , df2[cols].value_counts())
  print('******************')
```

```python
df2.head()
```

```python
#One hot encoding
pd.get_dummies(df2['gender'])
```

```python
non_num_data = df2[non_num].drop('Churn',axis='columns')
non_num_data.head()
```

```python
encoded_non_num = pd.get_dummies(non_num_data)
encoded_non_num.head()
```

```python
df2.head()
```

```python
df3 = df2.drop(non_num, axis='columns')
df3.head()
```

```python
X = pd.concat((df3,encoded_non_num),axis='columns')
X.head()
```

```python
Y = df2['Churn'].map({'No':0, 'Yes':1})
Y
```

```python
Y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=1)
X_res, Y_res = ros.fit_resample(X, Y)
```

**Split, Import algorithm and train the model**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X_res, Y_res, train_size = 0.80)

model2 = LogisticRegression()

model2.fit(xtrain,ytrain)
```

```python
print(model2.score(xtrain,ytrain))
print(model2.score(xtest,ytest))
```

```python
confusion_matrix(ytrain,model2.predict(xtrain))
```

```python
confusion_matrix(ytest , model2.predict(xtest))
```
